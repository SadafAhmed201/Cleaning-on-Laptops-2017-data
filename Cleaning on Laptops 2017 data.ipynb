# Data Cleaning

In the realm of data science, the journey from raw data to actionable insights often begins with meticulous data cleaning, a process akin to refining raw material into a polished gem. In my recent project focused on laptops data from 2017, I embarked on a thorough journey of data refinement, ensuring that the dataset was pristine and ready for analysis.

Task by Task Overview:

Throughout this project, I meticulously executed a series of data cleaning tasks, each designed to enhance the quality and usability of the dataset. Starting with standardizing column names and converting string values to numeric formats, I then delved into calculating averages and identifying outliers. From extracting GPU manufacturer information to handling missing values and ensuring consistency in OS categorization, every step was aimed at preparing a dataset that is not only accurate but also insightful.


```python
import pandas as pd
```

```python
laptops = pd.read_csv("laptops.csv", encoding = "latin-1")
```

```python
laptops
```

```python
 laptops.info()
```

```python
laptops.columns
```

```python
laptops.columns = laptops.columns.str.strip().str.replace("(", "").str.replace(")", "")\
.str.replace(" ", "_").str.lower().str.replace("Operating_System", "os")
```

```python
laptops
```

```python
laptops.rename(columns={"operating_system": "os", "operating_system_version": "os_version"}, inplace = True)
```

```python
laptops
```

```python
# Convert string screen sizes to numeric values
laptops['screen_size'] = laptops['screen_size'].str.strip('"').astype(float)
```

```python
# Calculate the average screen size
laptops['screen_size'].mean()
```

```python
 laptops.rename({"screen_size":"screen_size_inches"}, axis=1, inplace= True)
```

```python
laptops
```

```python
# Calculate average ram

# first add gb in column name
laptops.rename({"ram": "ram_gb"}, axis=1, inplace = True)

# now remove gb from column
laptops["ram_gb"] = laptops["ram_gb"] = laptops["ram_gb"].str.strip("GB").astype(int)
```

```python
laptops
```

```python
laptops["ram_gb"].mean()
```

```python
#now find the highest ram
laptops.sort_values(by="ram_gb", ascending=False)
```

```python
laptops["gpu_manufacturer"] = \
laptops["gpu"].str.split(n=1, expand = True) [0]
```

```python
laptops["gpu_manufacturer"].value_counts()
```

```python
laptops
```

```python
#find how many laptops from each OS

laptops["os"].value_counts()
```

```python
# replace Mac OS with macOS

laptops["os"].replace("Mac OS", "macOS", inplace = True )
```

```python
#find how many laptops from each OS

laptops["os"].value_counts()
```

```python
# count missing values

laptops.isnull().sum()
```

```python
# handle missing values in os_version
laptops.loc[laptops["os_version"].isnull(), "os"].value_counts()
```

```python
# add "Unknown" in missing values
laptops.loc[laptops["os_version"].isnull(), "os_version"] ="Unknown"
```
